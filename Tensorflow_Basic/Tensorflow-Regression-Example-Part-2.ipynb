```python
import tensorflow as tf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Define list of feature columnns - special estimator data type
# needs a key and shape

feature_col = [ tf.feature_column.numeric_column('x', shape=[1]) ]
```

```python
# create estimator model

estimator = tf.estimator.LinearRegressor(feature_columns=feature_col)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_data = np.linspace(0.0, 10.0 , 1000000)
```

```python
noise = np.random.randn(len(x_data))
```

```python
# y = mx + b
y_true = (0.8 * x_data) + 8 + noise
```

```python
x_df = pd.DataFrame(data=x_data , columns=["X Data"])
```

```python
y_df = pd.DataFrame(data= y_true, columns=["Y"])
```

```python
my_concat = pd.concat([x_df, y_df] , axis=1)
```

```python
x_train, x_eval, y_train, y_eval = train_test_split(x_data,y_true,
                                                    test_size=0.3,
                                                    random_state = 101)
```

```python
print(x_train.shape)
print(x_eval.shape)
```

```python
# data input function - numpy or pandas for format of data

input_func = tf.estimator.inputs.numpy_input_fn({'x': x_train}, y_train,
                                               batch_size=8, num_epochs=None,
                                               shuffle=True)
```

```python
# data input function - numpy or pandas for format of data

train_input_func = tf.estimator.inputs.numpy_input_fn({'x': x_train}, y_train,
                                               batch_size=8, num_epochs= 1000,
                                               shuffle = False)
```

```python
# data input function - numpy or pandas for format of data

eval_input_func = tf.estimator.inputs.numpy_input_fn({'x': x_eval}, y_eval,
                                               batch_size=8, num_epochs=None,
                                               shuffle=True)
```

```python
# train,evaluate,predict methods

estimator.train(input_fn = input_func, steps=1000)
```

```python
# evaluating aginst true input
train_metrics = estimator.evaluate(input_fn= train_input_func, steps=1000)
```

```python
eval_metrics = estimator.evaluate(input_fn = eval_input_func,steps = 1000)
```

```python
print("TRAINING DATA METRICS")
print(train_metrics)
```

```python
# we know if we are overfitting if we have low lost on train data but high on 
# eval data
print("EVAL METRICS")
print(eval_metrics)
```

```python
# now we are asking if we have new data aka x-values, what are the corresponding
# y-values based on our models
brand_new_data = np.linspace(0,10,10)
input_fn_predict = tf.estimator.inputs.numpy_input_fn({'x':brand_new_data},shuffle=False)
```

```python
# here we are predicting how the new data will do against our model
estimated_pred = list(estimator.predict(input_fn=input_fn_predict))
```

```python
estimated_pred
```

```python
predictions = []

for pred in estimator.predict(input_fn=input_fn_predict):
    predictions.append(pred['predictions'])
```

```python
predictions
```

```python
my_concat.sample(n=250).plot(kind='scatter' , x = "X Data", y ='Y')
plt.plot(brand_new_data,predictions, 'r*')
```
